```python
import pandas as pd

data = {
    'For backbone = Wide ResNet50': [
        'layer2, layer3',
        'layer1, layer2',
        'layer2, layer3, layer4'
    ],
    'AUROC': [0.8816, 0.9271, 0.8807],
    'Accuracy': [0.8375, 0.9125, 0.7625],
    'Recall': [0.975, 0.975, 0.9875],
    'Precision': [0.7647, 0.8667, 0.6810],
    'F1 Score': [0.8571, 0.9176, 0.8061],
}

df = pd.DataFrame(data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df.set_index('For backbone = Wide ResNet50').T, annot=True, cmap='coolwarm', linewidths=0.5)

# Set title
plt.title('PatchCore Model Performance (Heatmap)')

# Display the heatmap
plt.show()
```

```python
import matplotlib.pyplot as plt

# List of metric pairs to compare
comparisons = [
    ('Accuracy', 'Precision'),
    ('Accuracy', 'Recall'),
    ('Precision', 'Recall'),
    ('F1 Score', 'Accuracy')
]

# Create scatter plots for each pair of metrics
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

for i, (metric1, metric2) in enumerate(comparisons):
    axs[i].scatter(df[metric1], df[metric2], color='blue')
    axs[i].set_xlabel(metric1)
    axs[i].set_ylabel(metric2)
    axs[i].set_title(f'{metric1} vs {metric2}')

    # Annotate points with model configuration names
    for j, row in df.iterrows():
        axs[i].annotate(row['For backbone = Wide ResNet50'], (row[metric1], row[metric2]))

plt.tight_layout()
plt.show()
```
